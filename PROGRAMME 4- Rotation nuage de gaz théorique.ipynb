```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from scipy import *             
from pylab import *
import os

k=2*np.pi

#vitesse orbitale identique(*10^4km/s):
V1=17.39
V2=21.30
V3=30.12
V4=15.06
V5=13.47

#Rayons des nuages de gaz(kpc)
R1=30
R2=20
R3=10
R4=40



#Périodes orbitales
T1=R1/V1
T2=R2/V2
T3=R3/V3
T4=R4/V4


#angle continu
def theta(T):
    return 2*np.pi*np.linspace(0, T, 1000)/T

#Orbites des nuages de gaz:
x1=R1*np.cos(theta(T1))
y1=R1*np.sin(theta(T1))

x2=R2*np.cos(theta(T2))
y2=R2*np.sin(theta(T2))

x3=R3*np.cos(theta(T3))
y3=R3*np.sin(theta(T3))

x4=R4*np.cos(theta(T4))
y4=R4*np.sin(theta(T4))


#initialisation 
ti = 0 
tfin = 2*365*24
Npas = 2*365

#angle discret
def alpha(t,T):
    return k*t/T

#Positions Successives 
def Positions1(t): 
    x = R1*np.cos(alpha(t,T1))
    y = R1*np.sin(alpha(t,T1))
    return [x, y]

def Positions2(t):
    x = R2*np.cos(alpha(t,T2))
    y = R2*np.sin(alpha(t,T2))
    return [x, y]

def Positions3(t):
    x = R3*np.cos(alpha(t,T3))
    y = R3*np.sin(alpha(t,T3))
    return [x, y]

def Positions4(t):
    x = R4*np.cos(alpha(t,T4))
    y = R4*np.sin(alpha(t,T4))
    return [x, y]


# Construction d'une série d'images et de leur assemblage dans une animation 
for k in range(Npas): 
    t = ti + 1
    #Orbites des différents nuages de gaz:
    plt.plot(x1,y1,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x2,y2,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x3,y3,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x4,y4,color='blue',linewidth=1,linestyle='-.')

    #centre galactique:
    plt.plot(0,0,'o',color='red')
    #Positions au cours du temps des différents nuages de gaz:
    plt.plot(Positions1(t)[0], Positions1(t)[1], 'o', color='blue', label='nuage 1')
    plt.plot(Positions2(t)[0], Positions2(t)[1], 'o', color='blue', label='nuage 2')
    plt.plot(Positions3(t)[0], Positions3(t)[1], 'o', color='blue', label='nuage 3')
    plt.plot(Positions4(t)[0], Positions4(t)[1], 'o', color='blue', label='nuage 4')
   
    plt.axis('equal')
    plt.title('Simulation des vitesses théoriques')
    savefig('fichierTemp'+str('%02d' %k)+'.pdf') 
    clf()
    
plt.show()
```
