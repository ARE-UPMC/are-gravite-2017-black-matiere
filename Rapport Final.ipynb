Cependant, la principale difficulté est que nous nous situons à l'intérieur de la galaxie. Par conséquent, nos observations sont réalisées d'un point de vue interne, au niveau du système solaire lui même en déplacement par rapport au centre galactique. L'une des méthodes permettant d'accéder au mouvement entre un émetteur et un récepteur en mouvement l'un par rapport à l'autre selon une ligne de visée s'appuie sur l'effet Doppler. En effet, d'après cette méthode, l'étude du spectre de la lumière de l'émetteur comportera des décalages de longueur d'onde par rapport au spectre obtenu si les deux objets étaient au repos l'un par rapport à l'autre. Une longueur d'onde décalée vers le rouge (800nm) indique que la distance entre les deux objets augmente, l'émetteur s'éloigne du récepteur. Inversement, une longueur d'onde décalée vers le bleu (400nm) indique que la distance entre l'émetteur et le récepteur diminue, l'émetteur s'approche du récepteur. Nous étudierons par la suite le mouvement rotatoire de nuages de gaz autour du centre galactique, afin de modéliser et cartographier la Voie Lactée. ## Affichage des vitesses radiales des nuages de gaz:

Lors de l'observation télescopique de notre galaxie, il est possible de relever la présence de nuage de gaz le long de la ligne de visée. Nous pouvons alors observer la vitesse radiale, notée Vr par la suite, qui correspond à la projection de la vitesse réelle de ce nuage sur la ligne de visée à laquelle on soustrait la projection de la vitesse du Soleil. Nous avons tenter grâce au programme qui suit d'afficher la vitesse radiale d'un nuage de gaz en fonction de sa position. Pour cela, il est nécessaire de relier le centre galactique au nuage de point et au centre du Soleil. Par la suite, nous traçons la le segment perpendiculaire à la ligne de visée passant par le centre galactique. En reportant l'angle alpha compris entre le segment centre galactique/nuage et droite perpendiculaire à la ligne de visée sur la ligne de visée nous obtenons le vecteur vitesse radiale.


```python
#Affichage des vitesses radiales en fonction des coordonnées du nuage de gaz:
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
from __future__ import division
from scipy import *           
from pylab import *
import os

def cercle (rayon,pas):
    R=rayon
    alpha=0
    dalpha=pas
    while alpha<2*np.pi:
        alpha=alpha+dalpha
        X=R*cos(alpha)
        Y=R*sin(alpha)
        plot(X,Y, 'co',)
          
plt.axis("equal")       


def deuxcercles(r1,pas1,r2,pas2):
    C1=cercle(r1,pas1)
    plt.axis("equal")
    plt.plot(C1)
    C2=cercle(r2,pas2)
    plt.axis("equal")
    plt.plot(C2)
    plt.plot(0,0,'ro')

deuxcercles(3,np.pi/4,6,np.pi/8)

RS=6
RN=3
alpha=np.pi/4
beta=np.pi/8

def xsoleil(t):
    x = RS*np.cos(beta*t)
    return x

def ysoleil(t):
    y = RS*np.sin(beta*t)
    return y

def xnuage(t):
    x = RN*np.cos(alpha*t)
    return x

def ynuage(t):
    y = RN*np.sin(alpha*t)
    return y


def vitesserad(t):
    plt.axis([-10,10,-10,10])
    #coordonnées du Soleil et du nuage:
    x1=xsoleil(t)
    y1=ysoleil(t)
    x2=xnuage(t)
    y2=ynuage(t)
    #affichage des points au temps t:
    plt.plot(x1,y1,'ro')
    plt.plot(x2,y2,'ro')
    plt.plot(0,0,'co')
    #orbites du soleil et du nuage de gaz:
    theta = linspace(0, 2*pi,40)
    xc1 = 3*cos(theta)
    yc1 = 3*sin(theta)
    plot(xc1, yc1,'-.')
    xc2 = 6*cos(theta)
    yc2 = 6*sin(theta)
    plot(xc2, yc2,'-.')
    plt.axis('equal')
    #relier centre galactique/Soleil:
    plt.plot([0,x1],[0,y1],'-')
    #relier centre galactique/nuage:
    plt.plot([0,x2],[0,y2],'-')
    #droite soleil/nuage:
    a=x2-x1
    b=y2-y1
    #forme : m*x+p
    m1=b/a
    p1=y1-m1*x1
    x = np.linspace(-6, 6, 20)
    fx=m1*x+p1
    plt.plot(x,fx)
    #tracée de la perpandiculaire
    xp = np.linspace(-6, 6, 20)
    m2=-1/m1
    fxp=m2*xp
    plt.plot(xp,fxp)
    #distance jaune
    SN=sqrt((x1-0)**2+(y1-0)**2)
    #Xp=(-y1-m1*x1)/((b^2+a^2)/(a*b))
    #Yp=-(a/b)*(-y1-m1*x1)/((b^2+a^2)/(a*b))
    xP=((-b*a)/((a**2)+(b**2)))*(y1-x1*(b/a))
    yP=-(a/b)*xP
    #plt.plot(xP,yP,'ro')
    
    #point I
    #plt.plot(xP,y2,'ro')
    NI=sqrt((xP-x2)**2+(y2-y2)**2)
    IP=sqrt((xP-x2)**2+(yP-y2)**2)
    
    #point R
    #plt.plot((xP+NI),(yP+IP),'ro')
    
    #doite centre/soleil
    fx2=(x2/y2)*xp+y2-(x2/y2)*x2
    plt.plot(xp,fx2)
    
    #perpandiculaire passant par nuage:
    fx3=-(x2/y2)*xp+(y2+(x2/y2)*x2)
    plt.plot(xp,fx3)
    
vitesserad(2)
plt.show()

```

# Calcul des vitesses orbitales des nuages de gaz

L'ojectif de cette seconde partie est de tracer une courbe de rotation de la galaxie dans le quadrant I. Pour cela, nous allons créer un programme permettant d'obtenir directement la vitesse orbitale et le rayon galactique d'un nuage de gaz, connaissant la longitude galactique et la vitesse radiale de celui-ci. Nous allons trouver la vitesse radiale sur les spectres simulés, cependant, il peut y avoir de nombreux nuages le long de la ligne de visée. Sur les spectres, on observe alors de nombreux pics, chacun correspondant à un nuage le long de la ligne de visée. Il est néanmoins facile de déterminer le vecteur vitesse et le rayon du nuage ayant la plus grande vitesse radiale. Ce point est appelé le point tangentiel et en ce point le vecteur vitesse est projeté tout du long sur la ligne de visée. En relevant les vitesses radiales max pour des longitudes comprises entre 0° et 90°, nous obtiendrons assez de données pour tracer la courbe de rotation de la galaxie.


```python
#Programme permettant d'obtenir la Vitesse orbitale et le rayon du nuage de gaz au point tangentiel.

def vitessegaz(longitude, vitesserad):
    
    #vitesse du Soleil autour du centre galactique en (km/s):
    v0=220
    #Distance entre le soleil et le centre galactique(kpc, 1 pc=3.09*10**6m):
    R0=8.5
    #vitesses orbitales
    V=0
    #Distance entre le nuage et le centre galactique
    R=0
    
    V=vitesserad+v0*sin(math.radians(longitude))
    R=R0*sin(math.radians(longitude))
    
    return (R, V)
```

De la même façon que précédemment, notre courbe du quadrant I fait apparaitre une courbe de rotation plate, au-delàs d'un certain rayon, ici (environs 2 kpc), la vitesse orbitale des nuages de gaz ne dépend plus de rayon est tend à être constante autour d'une valeur comprise entre 210km/s et 220km/s. Si la vitesse orbitale des nuages de gaz est contante pour des rayons éloignés du centre galactique, on obtient alors une vitesse angulaire supérieure pour la matière au centre que pour la matière éloigné d'où l'apparition de bras spiraux. Nous allons par la suite créer une simulation de ces bras spiraux. Pour cela nous construisons des orbites plus ou moins grands et nous plaçons en orbite dessus des points avec une vitesse constante. Nous avons construit nos bras spiraux par assemblage d'images successives. En faisant varier le rayon de nos nuages de gaz (points bleus) et leurs vitesses angulaires à chaque instant t, nous obtenons une simulation temporelle de bras spiraux formés par les nuages de gaz.


```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from scipy import *             
from pylab import *
import os

k=2*np.pi

#vitesse orbitale identique(e01km/s):
V=21

#Rayons des nuages de gaz(kpc)
R1=10
R2=20
R3=30
R4=40
R5=50
R6=60
R7=70

#Périodes orbitales
T1=R1/V
T2=R2/V
T3=R3/V
T4=R4/V
T5=R5/V
T6=R6/V
T7=R7/V

#angle continu
def theta(T):
    return 2*np.pi*np.linspace(0, T, 1000)/T

#Orbites des nuages de gaz:
x1=R1*np.cos(theta(T1))
y1=R1*np.sin(theta(T1))

x2=R2*np.cos(theta(T2))
y2=R2*np.sin(theta(T2))

x3=R3*np.cos(theta(T3))
y3=R3*np.sin(theta(T3))

x4=R4*np.cos(theta(T4))
y4=R4*np.sin(theta(T4))

x5=R5*np.cos(theta(T5))
y5=R5*np.sin(theta(T5))

x6=R6*np.cos(theta(T6))
y6=R6*np.sin(theta(T6))

x7=R7*np.cos(theta(T7))
y7=R7*np.sin(theta(T7))

#initialisation 
ti = 0 
tfin = 2*365*24
Npas = 2*365

#angle discret
def alpha(t,T):
    return k*t/T

#Positions Successives 
def Positions1(t): 
    x = R1*np.cos(alpha(t,T1))
    y = R1*np.sin(alpha(t,T1))
    return [x, y]

def Positions2(t):
    x = R2*np.cos(alpha(t,T2))
    y = R2*np.sin(alpha(t,T2))
    return [x, y]

def Positions3(t):
    x = R3*np.cos(alpha(t,T3))
    y = R3*np.sin(alpha(t,T3))
    return [x, y]

def Positions4(t):
    x = R4*np.cos(alpha(t,T4))
    y = R4*np.sin(alpha(t,T4))
    return [x, y]

def Positions5(t):
    x = R5*np.cos(alpha(t,T5))
    y = R5*np.sin(alpha(t,T5))
    return [x, y]

def Positions6(t):
    x = R6*np.cos(alpha(t,T6))
    y = R6*np.sin(alpha(t,T6))
    return [x, y]

def Positions7(t):
    x = R7*np.cos(alpha(t,T7))
    y = R7*np.sin(alpha(t,T7))
    return [x, y]

# Construction d'une série d'images et de leur assemblage dans une animation 
for k in range(Npas): 
    t = ti + 1
    #Orbites des différents nuages de gaz:
    plt.plot(x1,y1,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x2,y2,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x3,y3,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x4,y4,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x5,y5,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x6,y6,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x7,y7,color='blue',linewidth=1,linestyle='-.')
    #centre galactique:
    plt.plot(0,0,'o',color='red')
    #Positions au cours du temps des différents nuages de gaz:
    plt.plot(Positions1(t)[0], Positions1(t)[1], 'o', color='blue', label='nuage 1')
    plt.plot(Positions2(t)[0], Positions2(t)[1], 'o', color='blue', label='nuage 2')
    plt.plot(Positions3(t)[0], Positions3(t)[1], 'o', color='blue', label='nuage 3')
    plt.plot(Positions4(t)[0], Positions4(t)[1], 'o', color='blue', label='nuage 4')
    plt.plot(Positions5(t)[0], Positions5(t)[1], 'o', color='blue', label='nuage 5')
    plt.plot(Positions6(t)[0], Positions6(t)[1], 'o', color='blue', label='nuage 7')
    plt.plot(Positions7(t)[0], Positions7(t)[1], 'o', color='blue', label='nuage 8')
    plt.axis('equal')
    plt.title('Simulation des bras spiraux éphèmères de la Voie lactée')
    savefig('fichierTemp'+str('%02d' %k)+'.pdf') 
    clf()





```

Pour tenter de relier les points en rotation entre eux, nous avons créé deux programmmes fournissant les coordonnées de ces nuages de gaz en fonction du temps. Les programmes sont les suivants:

## Comparaison avec le réalité: l'existence de la matière noire

En théorie, entre les deux objets en mouvement relatif étudiés seule l'attraction gravitationnelle agit. En appliquant la deuxième loi de Newton, on peut détermminer la vitesse des nuages de gaz: V=(G*M/R^2)1/2. D'après la formule de la gravitation, l'intensité de celle-ci décroit avec le rayon séparant les deux objets. Ainsi plus on s'éloigne du centre galactique plus la vitesse est théoriquement faible. Nous avons alors simuler ce que nous aurions dû obtenir en théorie.

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from scipy import *             
from pylab import *
import os

k=2*np.pi

#vitesse orbitale identique(e01km/s):
V1=40
V2=20
V3=10


#Rayons des nuages de gaz(kpc)
R1=10
R2=20
R3=40


#Périodes orbitales
T1=R1/V1
T2=R2/V2
T3=R3/V3


#angle continu
def theta(T):
    return 2*np.pi*np.linspace(0, T, 1000)/T

#Orbites des nuages de gaz:
x1=R1*np.cos(theta(T1))
y1=R1*np.sin(theta(T1))

x2=R2*np.cos(theta(T2))
y2=R2*np.sin(theta(T2))

x3=R3*np.cos(theta(T3))
y3=R3*np.sin(theta(T3))

#initialisation 
ti = 0 
tfin = 2*365*24
Npas = 2*365

#angle discret
def alpha(t,T):
    return k*t/T

#Positions Successives 
def Positions1(t): 
    x = R1*np.cos(alpha(t,T1))
    y = R1*np.sin(alpha(t,T1))
    return [x, y]

def Positions2(t):
    x = R2*np.cos(alpha(t,T2))
    y = R2*np.sin(alpha(t,T2))
    return [x, y]

def Positions3(t):
    x = R3*np.cos(alpha(t,T3))
    y = R3*np.sin(alpha(t,T3))
    return [x, y]


# Construction d'une série d'images et de leur assemblage dans une animation 
for k in range(Npas): 
    t = ti + 1
    #Orbites des différents nuages de gaz:
    plt.plot(x1,y1,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x2,y2,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x3,y3,color='blue',linewidth=1,linestyle='-.')

    #centre galactique:
    plt.plot(0,0,'o',color='red')
    #Positions au cours du temps des différents nuages de gaz:
    plt.plot(Positions1(t)[0], Positions1(t)[1], 'o', color='blue', label='nuage 1')
    plt.plot(Positions2(t)[0], Positions2(t)[1], 'o', color='blue', label='nuage 2')
    plt.plot(Positions3(t)[0], Positions3(t)[1], 'o', color='blue', label='nuage 3')
    
    plt.axis('equal')
    plt.title('Simulation des vitesses théoriques')
    savefig('fichierTemp'+str('%02d' %k)+'.pdf') 
    clf()
```

Pour les nuages de gaz les plus éloignés du centre galactique, les vitesses sont manifestement donjc plus grande dans en réalité que dans la théorie Newtonienne, c'est la preuve même de l'existence de la matière noire, qui permet d'expliquer ces différences de vitesses.


```python
#Calcul de la masse intérieure de la galaxie:

l,R=np.loadtxt('courberotation.txt',skiprows=1,usecols=(0,1),unpack=True)
l,R

def masse(lr):
    G=6.67*10**-11 # constance gravitaionnelle
    R=lr[0] # rayon(kpc)
    V=220*10**3 #m.s-1
    for i in range (len(lr)):
        if lr[i] > R:
            R=lr[i]
    M =((V**2)*R*(10**3)*3.086*10**16)/G
    return M
```

Bilan des forces exercées sur le système : 
- force gravitationelle : FG= GmM(<R)/R^2


Or on a ici un mouvement circulaire on peut donc en déduire que : 
- m*ac= m*V^2/R


Ainsi d'après la deuxième loi de Newton on peut égaliser ces fonctions : 
- m*ac=FG <=> GmM(<R)/R^2 = m*V^2/R 
-         <=> M(<R) = V^2*R/G


On peut donc en déduire que l'on peut determiner la masse de matière contenue dans un cercle de rayon r infèrieur à R en mesurant V. C'est pourquoi, pour mesurer la masse intèrieur de la voie lactée, il faut utilisé le plus grand rayon possible.
