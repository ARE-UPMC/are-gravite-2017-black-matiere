## Affichage des vitesses radiales des nuages de gaz:

Dans la partie qui suit, nous cherchons à faire un lien entre la vitesse radiale et les différents quadrants définis précédemment. L'hydrogène constitue 5% à 10% de la Voie lactée, ainsi l'étude de ses émissions radios nous permet grâce à l'effet Doppler évoqué lors de l'introduction, de mesurer la vitesse relative de chaque objet le long de la ligne de visée. En effet, la formule suivante permet de relier la fréquence observée du signal à la vitesse du gaz émetteur:

$$\frac{f-f0}{f0}=\frac{-v}{c}$$

Avec:
- $f0$ est la fréquence de la source
- $f-f0$ est le décalage de fréquence
- $f$ la fréquence de la raie observée

Ainsi, sachant que l'hydrogène atomique qui constitue ces nuages de gaz émet à une longueur d'onde de 21 cm, ce qui correspond à une fréquence de 1420 MHz, nous allons pouvoir déterminer la vitesse des nuages de gaz émetteurs grâce à la formule précédente.

Notre objectif pour cette première partie est de déterminer la vitesse radiale des nuages de gaz présents de long de la ligne de visée afin d’avoir accès au rayon galactique et à la vitesse orbitale de ce nuage. Pour cela, nous devons déterminer la différence entre la projection de la vitesse de ce nuage sur la ligne de visée et la projection de la vitesse du Soleil. La projection de la vitesse du Soleil $V0$ est facilement déterminée grâce à la trigonométrie: elle correspond à $$V0*sin(l)$$
Il est cependant plus compliqué de déterminer la projection de la vitesse du nuage. En effet, nous savons que la projection peut s'écrire en fonction de alpha, où alpha désigne l'angle entre le vecteur vitesse et la ligne de visée :$$V*cos(alpha)$$
Cependant, nous ne connaissons pas alpha, c'est pourquoi nous utilisons la méthode du point tangent qui facilite nos calculs. En effet, le point tangentiel est le point où le rayon(R) est minimum, il vaut alors :$$R0*sin(l)$$

Ainsi la vitesse radiale est maximale et projetée tout du long sur la ligne de visée. En étudiant la géométrie de la Galaxie, nous pouvons déterminer la distance séparant le centre Galactique et le point tangentiel de deux manières, ce qui nous permettra de lever l'inconnue sur cos(alpha). On obtient alors :
$$R0*sinl=R*cos(alpha)$$
<=> $$cos(alpha)=\frac{R0}{R}*sin(l)$$

Finalement, la vitesse radiale a pour valeur $$Vr = V*\frac{R0}{R}*sin(l)*V0*sin(l)$$
où:
- Vr : Vitesse radiale du nuage
- V : Vitesse du nuage de gaz
- R0 : Distance Soleil/centre Galactique
- l : Longitude galactique
- R : Distance entre le nuage et le centre Galactique


Notons que le signe de la vitesse radiale varie selon le quadrant. En effet, selon la position du nuage de gaz, le signe de sin(l) et du rapport $\frac{R0}{R}-1$  varie :
-	Dans le **quadrant I** :  si la distance entre le Soleil  et le centre Galactique R0 est supérieure à la distance entre le nuage et le centre galactique R alors $\frac{R0}{R}-1>0$  et $sin(l)$ pour l compris entre 0 °et 90° est positif donc la vitesse radiale est positive.
-	Dans le **quadrant II** : si la distance entre le Soleil et le centre Galactique est inférieure à la distance entre le nuage et le centre galactique $\frac{R0}{R}-1<0$. $Sin(l)$ est positif pour l compris entre 90°<l<180° donc la vitesse radiale est de signe négatif.
-	Dans le **quadrant III** : si la distance entre le Soleil et le centre Galactique est inférieure à celle entre le Soleil et le centre galactique $\frac{R0}{R}-1<0$. De plus $sin(l)<0$ pour 180°<l<270°, donc la vitesse radiale est positive.
-	Dans le **quadrant IV** : si la distance entre le Soleil et le centre Galactique est supérieure à celle entre le Soleil et le centre galactique $\frac{R0}{R}-1>0$ mais $sin(l)$ pour 270°<l<360° est négatif donc la vitesse radiale est négative.

Si la vitesse relative entre le nuage et le Soleil est négative comme dans les quadrants II et IV cela signifie d’après l’effet Doppler que le nuage s’éloigne du Soleil, le décalage entre les longueurs d’ondes est donc vers le bleu. Au contraire, dans les quadrants I et III, la vitesse relative entre les deux objets est positive, ils se rapprochent, le décalage des longueurs d’onde est donc vers le rouge. 

Ces caractéristiques de la vitesse relative en fonction du quadrant ont pu être observées lors de l’activité kinesthésique où l’élastique tendu entre l’objet observé : le nuage, et l’observateur : le Soleil s’étirait correspondant à un éloignement ou se détendait correspondant à un rapprochement entre les deux objets.

##### Explication du programme suivant:

Tout d'abord la fonction "cercle" qui est la première fonction définie a pour but de tracer les trajectoires du nuage de gaz et du Soleil. Nous souhaitons faire avancer nos objets avec une vitesse constante, c'est pourquoi il nous faut définir des pas de temps correspondant aux positions prises par les deux astres aux cours du temps. Nous incrémentons donc une variable alpha à chaque tour de boucle avec un pas noté "delta". L'angle étant maintenant défini il ne nous reste qu'à définir les coordonnées de chaque points connaissant la distance R au centre et la direction donnée par alpha. Nous trouvons que : $x=R*cos(alpha)$ et $y=R*sin(alpha)$

Ensuite nous avons programmé quatre fonctions permettant de calculer les coordonnées x et y du Soleil et du nuage de gaz à un instant t. En supposant que la vitesse du nuage de gaz et celle du Soleil sont identiques, les deux objets ont la même vitesse mais une vitesse angulaire différente. Le rayon du Soleil étant de 8.5kpc, nous avons choisi arbitrairement un nuage de gaz avec un rayon deux fois inférieur à celui du Soleil soit 4.25 kpc du centre galactique, ainsi la vitesse angulaire du nuage de gaz est deux fois supérieure à celle du Soleil.

Il nous reste à determiner la vitesse radiale à un instant t, nous connaissons sa norme avec la formule $Vr = V*\frac{R0}{R}*sin(l)*V0*sin(l)$, cependant il nous faut trouver sa direction. Tout d'abord nous devons relier les objets au centre Galactique, ce qu'il est possible de faire en reliant simplement les coordonnées des positions des deux astres au cours du temps.

Ensuite nous cherchons à tracer la perpendiculaire à la ligne de visée passant par le centre de la Galaxie afin de reproduire l'angle entre cette perpendiculaire et la droite centre/nuage sur la ligne de visée. Il s'agit de l'angle nommé alpha précédemment.

Nous savons que l'équation d'une droite passsant par des points $(x1,y1)$, $(x2,y2)$ s'écrit sous la forme **mx+p** avec m le coefficient directeur et p l'ordonnée à l'origine. Nous avons trouvé que m vaut $$m=\frac{y2-y1}{x2-x1}$$

L'ordonnée à l'origine est $$p=y1-m*x1$$ Avec le tracé de cette droite, on obtient l'angle alpha à reproduire sur la ligne de visée.

Enfin, nous avons tracé la perpendiculaire à la droite centre/nuage passant par le nuage. Ainsi par des propriétés géométriques, nous avons donc reproduit l'angle alpha qui nous donne la direction de la vitesse radiale. La dernière condition "if" permet d'obtenir un vecteur radial orienté dans la bonne direction selon le quadrant où se situe le nuage.


```python
#Programme 1- Vitesse radiale en annexe
#Affichage des vitesses radiales en fonction des coordonnées du nuage de gaz:

def cercle (rayon,pas):
    R=rayon
    alpha=0
    delta=pas
    while alpha<2*np.pi:
        alpha=alpha+delta
        X=R*cos(alpha)
        Y=R*sin(alpha)
        plot(X,Y, 'co',)
          
plt.axis("equal")       
plt.plot(cercle(4.25,np.pi/4), cercle(8.5,np.pi/8))

RS=8.5
RN=4.25
alpha=np.pi/4
beta=np.pi/8

#Coordonnnées du Soleil et du nuage de gaz:
def xsoleil(t):
    x = RS*np.cos(beta*t)
    return x

def ysoleil(t):
    y = RS*np.sin(beta*t)
    return y

def xnuage(t):
    x = RN*np.cos(alpha*t)
    return x

def ynuage(t):
    y = RN*np.sin(alpha*t)
    return y

def vitesserad(t):
    plt.axis([-5,5,-5,5])
    plt.axis('equal')
    
    #coordonnées du Soleil et du nuage de gaz à l'instant t:
    x1=xsoleil(t)
    y1=ysoleil(t)
    x2=xnuage(t)
    y2=ynuage(t)
    
    #affichage des astres à l'instant t:
    plt.plot(x1,y1,'ro',color="green")
    plt.plot(x2,y2,'ro')
    plt.plot(0,0,'co')

    #Relier centre galactique/Soleil:
    plt.plot([0,x1],[0,y1],'-')
    
    #Relier centre galactique/nuage:
    plt.plot([0,x2],[0,y2],'-')
    
    #Droite Soleil/nuage:
    a=x2-x1
    b=y2-y1
    
    #forme : m*x+p
    m1=b/a
    p1=y1-m1*x1
    x = np.linspace(-10, 10, 20)
    fx=m1*x+p1
    plt.plot(x,fx)
    
    #Tracé de la perpendiculaire:
    xp = np.linspace(-6, 6, 20)
    xpr = np.linspace(-8, 8, 20)
    m2=-1/m1
    fxp=m2*xpr
    plt.plot(xpr,fxp)
    
    #distance jaune
    SN=sqrt((x1-0)**2+(y1-0)**2)
    #Xp=(-y1-m1*x1)/((b^2+a^2)/(a*b))
    #Yp=-(a/b)*(-y1-m1*x1)/((b^2+a^2)/(a*b))
    xP=((-b*a)/((a**2)+(b**2)))*(y1-x1*(b/a))
    yP=-(a/b)*xP
    #plt.plot(xP,yP,'ro')
    
    #point I
    #plt.plot(xP,y2,'ro')
    NI=sqrt((xP-x2)**2+(y2-y2)**2)
    IP=sqrt((xP-x2)**2+(yP-y2)**2)
    
    #point R
    #plt.plot((xP+NI),(yP+IP),'ro')
    
    #droite centre/soleil
    fx2=(x2/y2)*xp+y2-(x2/y2)*x2
    plt.plot(xp,fx2)
    
    #perpendiculaire passant par le nuage:
    fx3=-(x2/y2)*x+(y2+(x2/y2)*x2)
    plt.plot(x,fx3,'-',color="pink")
    
    #FLECHE :
    xf=xnuage(t+1)
    yf=ynuage(t+1)
    
    if (t>=0 and t<=2) or (t>=6 and t<=7) :
        mf2=-(x2/y2)
        mf1=1  
    else:
        mf2=(x2/y2)
        mf1=-1
 
    ax = plt.axes()
    ax.arrow(x2, y2, mf1, mf2, head_width=0.2, head_length=0.5, fc='k', ec='k')
    
    annotate("Soleil", xy=(x1,y1), xycoords="data",
                  xytext=(x1-3,y1+3),
                  va="top", ha="center",
                  bbox=dict(boxstyle="round", fc="w"),
                  arrowprops=dict(arrowstyle="->"))
    annotate("Nuage", xy=(x2,y2), xycoords="data",
                  xytext=(x2-8,y2),
                  va="top", ha="center",
                  bbox=dict(boxstyle="round", fc="w"),
                  arrowprops=dict(arrowstyle="->"))
    


vitesserad(3)
plt.show()
```

##### Explication du programme "Rotation du nuage de gaz et Soleil":
Dans ce second programme nous voulons simplement montrer le mouvement du Soleil et du nuage de gaz par rapport au centre galactique. Pour cela de la même manière que précédemment, nous traçons les orbites des deux astres. A l'aide de la fonction "animate" nous créons deux points en mouvement qui sont en orbite avec une vitesse angulaire respective de $pi/12$ et $pi/24$. Nous avons modifié les vitesses angulaires par rapport à la modélisation précédente afin d'avoir des vitesses adéquates à l'observation.

```python
#Rotation du nuage de gaz et du Soleil par rapport au centre galactique.
from matplotlib import animation

def deuxcercles(r1,pas1,r2,pas2):
    C1=cercle(r1,pas1)
    plt.axis("equal")
    plt.plot(C1)
    C2=cercle(r2,pas2)
    plt.axis("equal")
    plt.plot(C2)
    plt.plot(0,0,'ro')

fig = plt.figure() 
fig.set_dpi(100) 
fig.set_size_inches(7, 6.5) 
ax = plt.axes(xlim=(-10, 10), ylim=(-10, 10))
 
size = 0.3
xcenter = 0
ycenter = 0
radius1 = 4.25
radius2= 8.5
 
patch1 = plt.Circle((0,0), size, fc='y') 
patch2 = plt.Circle((0,0), size, fc='r') 

def init(): 
    ax.add_patch(patch1)
    ax.add_patch(patch2)
    return patch1, patch2 

def animate(t): 
    x1, y1 = patch1.center 
    x1 = xcenter + radius1 * np.sin((t*np.pi/12))
    y1 = ycenter + radius1 * np.cos((t*np.pi/12))
    patch1.center = (x1, y1)
    x2, y2 = patch2.center 
    x2 = xcenter + radius2 * np.sin((t*np.pi/24))
    y2 = ycenter + radius2 * np.cos((t*np.pi/24))
    patch2.center = (x2, y2)
    return patch1, patch2
 
anim = animation.FuncAnimation(fig, animate,  
                               init_func=init,  
                               frames=1000,  
                               interval=20, 
                               blit=True) 
 


deuxcercles(4.25,np.pi/12,8.5,np.pi/24)
plt.axis("equal")
plt.show()
```

# Calcul des vitesses orbitales des nuages de gaz

Nous connaissons maintenant, grâce aux observations télescopiques de l’interface EU HOU, la vitesse radiale du point tangentiel pour chaque longitude comprise entre 0° et 90° (quadrant I). Nous avons donc construit un programme permettant de connaître la vitesse orbitale et le rayon du nuage de gaz en entrant la longitude et la vitesse radiale grâce aux formules suivantes : $$R=R0*sin(l)$$
$$V=Vrmax+V0*sin(l)$$

où:
- $Vrmax$ correspond à la vitesse radiale au point tangent.
- $V$: Vitesse du nuage de gaz
- $R0$ : Distance Soleil/centre Galactique 8.5kpc
- $l$: Longitude galactique
- $R$: Distance entre le nuage et le centre Galactique
- $V0$: Vitesse de rotation du Soleil: 220km/s

En effet, le rayon R évoqué correspond au côté opposé à l'angle dans le triangle formé par le centre galactique, le Soleil et le nuage de gaz (ligne de visée). Ainsi l'hypothénuse correspond au rayon du Soleil R0 et l'angle formé par R0 et la ligne de visée correspond à la longitude. La première formule est dès lors évidente. La seconde est liée à la définition de la vitesse radiale. Si celle-ci est la différence entre les projections des vitesses du Soleil et du nuage, nous pouvons en déduire que la vitesse du nuage est la somme de la vitesse radiale et de la projection du Soleil, soit $V0*sin(l)$, d'où la deuxième formule.

Avec les résultats obtenus, nous pouvons tracer une courbe de rotation de la Galaxie pour le quadrant I représentant l’évolution de la vitesse orbitale des nuages de gaz en fonction de la distance centre Galactique/nuage de gaz. Cette courbe de rotation nous permettra par la suite de remonter à la répartition de la masse dans la Galaxie grâce à la deuxième loi de Newton qui lie l'accélération et la masse.


##### Explication du programme de calcul de la vitesse radiale:
Le programme suivant ne nécessite que très peu d'explication puisqu'il s'agit d'une application de la formule de la vitesse radiale ci-dessus. Il a simplement été nécessaire de rentrer les constantes R0, V0, et de convertir les angles en radians grâce à la fonction "math.radians".

```python
#Programme permettant d'obtenir la Vitesse orbitale et le rayon du nuage de gaz au point tangentiel.

def vitessegaz(longitude, vitesserad):
    
    #vitesse du Soleil autour du centre galactique en (km/s):
    v0=220
    #Distance entre le soleil et le centre galactique(kpc, 1 pc=3.09*10**6m):
    R0=8.5
    #vitesses orbitales
    V=0
    #Distance entre le nuage et le centre galactique
    R=0
    
    V=vitesserad+v0*sin(math.radians(longitude))
    R=R0*sin(math.radians(longitude))
    
    return (R, V)
```

## Carte de la Galaxie:

Désormais que nous savons comment tournent les nuages de gaz au sein de la Voie lactée, nous aimerions représenter la position de tous les nuages dans la Galaxie afin de la cartographier. Auparavant nous ne prenions en compte que la vitesse radiale du point tangent afin de connaître le comportement rotatoire des nuages de gaz. Nous allons maintenant prendre en compte les vitesses radiales de tous les pics présents sur les spectres. Avec ces vitesses radiales nous pouvons retrouver la distance du nuage au centre Galactique. Cependant, il serait trop compliqué de résoudre l’équation donnant R pour une valeur de Vr et de l, c’est pourquoi nous supposons que V=V0. Le rayon obtenu est alors donné par la formule suivante :
$$ R=\frac{R0*v0*sin(l)}{V0*sin(l)+Vr}$$
Il est nécessaire de remarquer que :
-	si nous observons pour R inférieur à R0 dans les quadrants I et IV, il y a 2 positions possibles de R  correspondant aux points d’intersection M et M’ de la ligne de visée ST et du cercle intérieur. 
-	Si nous observons à R supérieur à R0 dans les quadrants II et III, la position du nuage de gaz est unique. 

En effet, dans le triangle CSM où C est le centre Galactique, S le centre du Soleil et M la position du nuage de gaz, nous pouvons écrire la relation suivante :
$$R^2=R0^2+r^2-2*R0*r*cos(l)$$
En résolvant le trinôme, nous obtenons donc :
$$R=+-(R^2-R0^2*sin^2(l))^(1/2)+R0*cos(l)$$

Nous conserverons seulement les valeurs positives de ces rayons, les valeurs négatives de r sont éliminées car elles n’ont pas de réalité physique.

Nous savons que les coordonnées galactiques ont pour origine le centre du Soleil. Nous devons donc trouver la distance du nuage au centre du Soleil. En connaissant la direction grâce à la longitude, nous serons capables de dessiner une carte de la Galaxie. Avec les relations trigonométriques, nous pouvons déterminer les coordonnées X et Y du rayon reliant le centre du Soleil à la position du nuage de gaz. Cependant, la longitude et l’angle theta (qui nous servira pour repérer l'angle entre l'axe des abscisses et le rayon du nuage de gaz) ne sont pas définis de la même manière.  Afin de trouver theta il faut soustraire 90° à la longitude, car la longitude zéro est « au pôle sud », comme schématisée sur la première figure des quadrants galactiques. Nous avons maintenant la direction et les coordonnées du points représentant le nuage de gaz.

$$X=r*cos(theta)$$
$$Y=r*sin(theta)$$

## Comparaison avec le réalité: l'existence de la matière noire


Nous savons que la vitesse est étroitement liée à la masse de la Galaxie. En effet, la seule force exercée entre les deux objets : nuage et centre Galactique est la force gravitationnelle. Par application de la deuxième loi de Newton nous obtenons : 
$$ m*A=\frac{G*M*m}{R^2} $$
où:
- m est la masse du nuage de gaz
- A est l'accélération du nuage de gaz
- M est la masse de la Galaxie estimée à 4.2*1041 kg 

Nous pouvons estimer la quantité de masse dans la Galaxie en regardant la localisation de la plus grande quantité de lumière émise par celle-ci. Une grande quantité de lumière, est signe d'un nombre important d’étoiles et donc de plus de masse. Lors de l'observation de la Voie lactée, nous trouvons que le centre de la Galaxie émet plus de lumière que les régions externes, ainsi nous pouvons supposer que la masse de la Galaxie est concentrée en son centre.

Si l’on suppose que la masse principale de la Galaxie se situe en son centre et que la masse des nuages de gaz est négligeable par rapport à la masse de la Galaxie, nous pouvons par conséquent affirmer que le centre de masse de la Galaxie est proche du centre Galactique. Ainsi, l’accélération centrifuge de la vitesse radiale des nuages de gaz vaut $$\frac{V^2}{R}=\frac{G*M}{R^2}$$
Nous obtenons donc notre lien entre la vitesse et la force gravitationnelle : 
$$V=(\frac{G*M}{R})^{1/2}$$
Avec cette formule nous sommes capable de tracer une courbe de rotation képlérienne qui modélise l'évolution des vitesses que nous devrions obtenir en théorie. En considérant la masse de la galaxie constante : 4,2.10^41 kg et G la constante gravitationnelle universelle constante à 6,67*10^-11N⋅m2⋅kg-2. On peut donc en déduire l’allure de la courbe : la vitesse diminue quand le rayon augmente avec une portée de 1/R. En théorie, on obtiendrait donc une courbe de rotation similaire à celle ci-dessous et non une courbe plate comme obtenue précédemment.

```python
# PROGRAMMME 4 en annexe:

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
from scipy import *             
from pylab import *
import os

k=2*np.pi

#vitesse orbitale identique(*10^4km/s):
V1=17.39
V2=21.30
V3=30.12
V4=15.06
V5=13.47

#Rayons des nuages de gaz(kpc)
R1=30
R2=20
R3=10
R4=40

#Périodes orbitales
T1=R1/V1
T2=R2/V2
T3=R3/V3
T4=R4/V4


#angle continu
def theta(T):
    return 2*np.pi*np.linspace(0, T, 1000)/T

#Orbites des nuages de gaz:
x1=R1*np.cos(theta(T1))
y1=R1*np.sin(theta(T1))

x2=R2*np.cos(theta(T2))
y2=R2*np.sin(theta(T2))

x3=R3*np.cos(theta(T3))
y3=R3*np.sin(theta(T3))

x4=R4*np.cos(theta(T4))
y4=R4*np.sin(theta(T4))


#initialisation 
ti = 0 
tfin = 2*365*24
Npas = 2*365

#angle discret
def alpha(t,T):
    return k*t/T

#Positions Successives 
def Positions1(t): 
    x = R1*np.cos(alpha(t,T1))
    y = R1*np.sin(alpha(t,T1))
    return [x, y]

def Positions2(t):
    x = R2*np.cos(alpha(t,T2))
    y = R2*np.sin(alpha(t,T2))
    return [x, y]

def Positions3(t):
    x = R3*np.cos(alpha(t,T3))
    y = R3*np.sin(alpha(t,T3))
    return [x, y]

def Positions4(t):
    x = R4*np.cos(alpha(t,T4))
    y = R4*np.sin(alpha(t,T4))
    return [x, y]


# Construction d'une série d'images et de leur assemblage dans une animation 
for k in range(Npas): 
    t = ti + 1
    #Orbites des différents nuages de gaz:
    plt.plot(x1,y1,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x2,y2,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x3,y3,color='blue',linewidth=1,linestyle='-.')
    plt.plot(x4,y4,color='blue',linewidth=1,linestyle='-.')

    #centre galactique:
    plt.plot(0,0,'o',color='red')
    #Positions au cours du temps des différents nuages de gaz:
    plt.plot(Positions1(t)[0], Positions1(t)[1], 'o', color='blue', label='nuage 1')
    plt.plot(Positions2(t)[0], Positions2(t)[1], 'o', color='blue', label='nuage 2')
    plt.plot(Positions3(t)[0], Positions3(t)[1], 'o', color='blue', label='nuage 3')
    plt.plot(Positions4(t)[0], Positions4(t)[1], 'o', color='blue', label='nuage 4')
   
    plt.axis('equal')
    plt.title('Simulation des vitesses théoriques')
    savefig('fichierTemp'+str('%02d' %k)+'.pdf') 
    clf()
```

### Comment expliquer cette différence entre la réalité et la théorie ?
La seule hypothèse que nous ayons faite pour réaliser cette courbe théorique est que la masse de la Galaxie soit concentrée en son centre. Or cela ne conduit pas à la courbe obtenue avec nos observations réelles. On peut donc remettre en cause cette hypothèse. Où se situe donc la masse de la Galaxie ? Cet écart entre théorie et réalité semble être une voie indirecte de montrer l’existence de la matière noire dans la Galaxie. En effet, si sa masse n’est pas concentrée en son centre, on peut supposer qu’il existe de la masse localisée ailleurs et que l’on ne peut pas voir.

## Estimation de la répartition de masse de la Voie lactée

On peut donc estimer la répartition de la masse de la Galaxie en fonction du rayon. En supposant cette fois ci que la masse de la Galaxie soit distribuée sphériquement autour de son centre et non plus concentrée en celui-ci, nous pouvons calculer la masse intérieure pour un rayon donné.  A l’aide du théorème de Jeans, d’après lequel la masse en dehors d’un rayon n’affecte pas la vitesse de ce rayon, et sachant que la matière contenue dans ce même rayon à le même mouvement rotatoire que si la masse était concentrée au centre, nous pouvons connaître la masse de la Galaxie en fonction du rayon par la formule suivante :
$$\frac{V^2}{R}= \frac{G*M(<R)}{R}$$

Le programme suivant calcule la masse intérieur pour une liste de rayon donnée.


```python
#Calcul de la masse intérieure de la galaxie:

l,R=np.loadtxt('courberotation.txt',skiprows=1,usecols=(0,1),unpack=True)
l,R

def masse(lr):
    G=6.67*10**-11 # constance gravitaionnelle
    R=lr[0] # rayon(kpc)
    V=220*10**3 #m.s-1
    for i in range (len(lr)):
        if lr[i] > R:
            R=lr[i]
    M =((V**2)*R*(10**3)*3.086*10**16)/G
    return M
```

Lorsque nous traçons la courbe de l'évolution de la masse intérieure à un rayon en fonction de la distance séparant le nuage de gaz du centre Galactique, nous observons que celle-ci augmente linéairement. Chaque 5kpc, elle augmente de + 2,23932E+40 kg ainsi, si le rayon double, la masse intérieure double aussi, comme nous le voyons sur la courbe ci-dessus. Nous pouvons donc en conclure que la masse  de la Voie lactée  est distribuée spériquement autour du centre galactique, sous forme de matière sombre. Ainsi, nous avons pu modéliser la répartion de la masse de la Galaxie sur le schéma suivant. Nous remarquerons que la valeur de la masse associée à une couleur dans la légende comprend l'intégralité du cercle soit la masse des cercles de rayon inférieur. Nous avons donc l'explication à la vitesse constante au delà d'un certain éloignement du centre galactique. En effet si la masse n'est plus constante mais augmente avec le rayon nous obtenons une compensation de la décroissance 1/R dans la formule suivante: $$V=(\frac{G*M}{R})^{1/2}$$
La vitesse des nuage de gaz est donc constante. De plus nous pouvons avoir accès à la masse totale de la Galaxie à 50 kpc qui correspond à la taille de la Voie lactée, la masse de la Galaxie est 1,11966E+42kg soit 5,59829E+11 unité de masse solaire. Nous savons que la masse d'étoiles dans la Galaxie est de 2E11 unité de masse solaire ce qui est différent de la masse totale dela Galaxie. Il existe donc une masse indétectable autre que celle des étoiles, c'est la matière noire.
